```python
print(df_survey['Age'].value_counts())
```

1. Створюємо нову колонку в df_survey, називаємо її 'UsesPython' (True/False) - чи використовує людина Python.
2. .str - працюємо з колонкою 'LanguageHaveWorkedWith', як з текстом.
3. Перевіряємо .contains — кожну клітинку — чи є всередині потрібне слово 'Python', щоб  не  було помилки - na=False.
4. Розкладаємо людей по групах за віком python_by_age. Групуємо по віку та по користуванням Python.
5. .agg - одночасно рахуємо - Всього='count', Python_розробників='sum'.
6. Для кожної групи рахуємо всього людей і скільки з них Python розробники.
7. .agg - одночасно рахуємо - рбимо нові стовпчики - Всього ='count', Python_розробників ='sum'.
8. Рахуємо відсоток Python розробників в кожній групі. Для цього добавляємо новий стовпчик 'Відсоток_Python' до датасету python_by_age через python_by_age['Python_розробників'] / python_by_age['Всього'] * 100. (.round(1) - округлюємо до 1 знака
9. Сортуємо від найбільшого відсотку до найменшого.
10. Виводимо результат.

```python
df_survey['UsesPython'] = df_survey['LanguageHaveWorkedWith'].str.contains('Python', na=False)
python_by_age = df_survey.groupby('Age')['UsesPython'].agg(
    Всього ='count',
    Python_розробників ='sum'
)

python_by_age['Відсоток_Python'] = (python_by_age['Python_розробників'] / python_by_age['Всього'] * 100).round(1)
python_by_age = python_by_age.sort_values('Відсоток_Python', ascending=False)
print(python_by_age)
```

Рахуємо 75-й перцентиль компенсації та виводимо на екран через  f-рядок, де :,.0f - форматуємо число, : — означає "починається форматування", , — розділяє тисячі комою, .0 — кількість знаків після крапки - в нас 0f.  
2. Фільтруємо за двома критеріями - висока компенсація І віддалена робота - high_paid_remote.
3. Знаходимо найпопулярніші індустрії - .value_counts() — рахує частоту кожної індустрії і сортує за спаданням.

```python
percentile_75 = df_survey['ConvertedCompYearly'].quantile(0.75)
print(f"75-й перцентиль компенсації: ${percentile_75:,.0f}")

high_paid_remote = df_survey[
    (df_survey['ConvertedCompYearly'] > percentile_75) & # & — оператор "І" для двох умов одночасно
    (df_survey['RemoteWork'] == 'Remote')
]
print(f"Кількість відфільтрованих респондентів: {len(high_paid_remote)}")

top_industries = high_paid_remote['Industry'].value_counts().head(10)

print("\nТоп-10 індустрій серед високооплачуваних віддалених працівників:")
print(top_industries)
```

2. Далі йдуть індустрії: від 190 до 138 - Fintech, Healthcare, Other:, Internet, Telecomm or Information Services. І тут Fintech та Healthcare активно залучають працівників на віддалену роботу - 190-188.
